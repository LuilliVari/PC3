#PARTE I

Extraemos nuestros links del github, los cuales estaban previamente en el github desktop.

```python
education_expenditures_Link="https://github.com/LuilliVari/PC3/raw/refs/heads/main/Education%20expenditures.csv"
electricity_Link="https://github.com/LuilliVari/PC3/raw/refs/heads/main/Electricity%20-%20installed%20generating%20capacity.csv"
energy_consumption_Link="https://github.com/LuilliVari/PC3/raw/refs/heads/main/Energy%20consumption%20per%20capita.csv"
reserves_foreign_Link="https://github.com/LuilliVari/PC3/raw/refs/heads/main/Reserves%20of%20foreign%20exchange%20and%20gold.csv"
```

#Parte 4

Clasificamos cada una de estas variables en cinco intervalos para tener un merge final con todos los datos puestos correctamente.

```python
import pandas as pd

# Etiquetas para los intervalos
labels = ['Very Low', 'Low', 'Medium', 'High', 'Very High']

# Crear nuevas columnas por separado
gdp_intervals = pd.cut(Final_merge['%_of_GDP'], bins=5, labels=labels, include_lowest=True)
kw_intervals = pd.cut(Final_merge['kW'], bins=5, labels=labels, include_lowest=True)
btu_intervals = pd.cut(Final_merge['Btu/person'], bins=5, labels=labels, include_lowest=True)

# Unirlas al DataFrame original
Final_merge['GDP_interval'] = gdp_intervals
Final_merge['kW_interval'] = kw_intervals
Final_merge['Btu_interval'] = btu_intervals

# Mostrar resultado
Final_merge
```
